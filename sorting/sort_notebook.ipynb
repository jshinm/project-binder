# Sorting Algorithm Implementations

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dset = np.random.randint(low=0, high=10000, size=5000, dtype=int)
idx = np.arange(start=0, stop=len(dset))
```

```python
def plot(arr=None):
    if arr is not None:
        fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(5*2,5))
        ax[0].scatter(x=idx, y=dset, s=2)
        ax[1].scatter(x=idx, y=arr, s=2)
    else:
        fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(5*1,5))
        ax.scatter(x=idx, y=dset, s=2)
```

```python
plot()
```

## Bubble Sort

```python
def bubbleSort(dat):
    arr = dat.copy()
    swapped = False

    for i in range(len(arr)-1,0,-1):
        for j in range(i):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swapped = True
        if swapped:
            swapped = False
        else:
            break
    return arr
```

```python
out = bubbleSort(dset)
```

```python
plot(out)
```
